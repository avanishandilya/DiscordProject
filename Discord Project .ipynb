# Finding the Average Number of Messages overall 

```python
# creating pandas series of the years 
year_2018 = df[2018][0:22]
year_2019 = df[2019][0:22]
year_2020 = df[2020][0:22]
year_2021 = df[2021][0:22]
year_2022 = df[2022][0:22]
year_2023 = df[2023][0:22]
total = df["Total"]
```

```python
# calculating the averages of each year 
mean_18 = year_2018.mean()
mean_19 = year_2019.mean()
mean_20 = year_2020.mean()
mean_21 = year_2021.mean()
mean_22 = year_2022.mean()
mean_23 = year_2023.mean()
```

```python
# making a list of the average number of messages sent in a year
average_list = []
average_list.append(mean_18)
average_list.append(mean_19)
average_list.append(mean_20)
average_list.append(mean_21)
average_list.append(mean_22)
average_list.append(mean_23)
```

```python
print(average_list)
```

```python
header = df[:0]
```

```python
years = []
for y in header: 
    if type(y) == int: 
        years.append(y)
        
print(years)
```

```python
idx = 0 
for y in years: 
    average_list[idx] = round(average_list[idx], 2)
    print(str(y) + ": " + str(average_list[idx]))
    idx += 1
```

```python
print(average_list)
```

The following average is for the average messages sent through the years by all people. This average represents about the average of how many messages could be sent in a given year. 

```python
#for loop to sum up all the averages
average_sum = 0
length = len(average_list)
for x in average_list: 
    average_sum += x
average_message_overall = average_sum/length
print(average_message_overall)
```

```python
names = df['Name '].tolist()
```

```python
i = 0
for n in names:
    i += 1
    frame = df[df['Name '] == n]
    lst = [int(frame[2018]), int(frame[2019]), int(frame[2020]), int(frame[2021]), int(frame[2022]), int(frame[2023])]
    plt.style.use('default')
    fig, ax = plt.subplots(figsize = (16, 8))
    plt.subplots_adjust(bottom=0.2)
    csfont = {'fontname':'Brush Script MT'}
    ax = plt.gca()
    ax.set_xticks(years)
    style = {'family': 'serif', 'size': 24}
    l = n +'\'s Number of Messages Through the Years'
    plt.title(l, fontdict = style)
    
    style = {'family': 'serif', 'size': 20}
    plt.xlabel("Years", fontdict = style, labelpad = 10)
    maximum = max(lst) + 1000
    ax.set_ylim(0, maximum)
   
    plt.ylabel("Number of Messages", fontdict = style, labelpad = 10)
    plt.plot(years, lst, color = 'purple', marker = 'o', linewidth = 3)

    plt.grid()
    plt.savefig('Person' + str(i) + ".png")
    
   
    
plt.close()
```
